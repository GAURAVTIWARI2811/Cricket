```python
with open("t20_json_files/t20_wc_batting_summary.json") as k:
    j = json.load(k)

    records_batt = []
    for rec in j:
        records_batt.extend(rec["battingSummary"])
```

```python
df_batt = pd.DataFrame(records_batt)
```

```python
df_batt.head(11)
```

```python
df_batt['OUT/NOTOUT'] = df_batt.dismissal.apply(lambda x: "Out" if len(x)>0 else 'Not Out')
df_batt.head()
```

```python
df_batt.drop(['dismissal'],axis = 1,inplace = True)
```

```python
df_batt.head(11)
```

```python
df_batt['batsmanName'] = df_batt['batsmanName'].apply(lambda x: x.replace("â€",""))
df_batt.head(25)
```

```python
df_match.head()
```

```python
row_match_ID = {}

for index,row in df_match.iterrows():
    key1 = row['team1']+" "+'Vs'+" " + row['team2']
    key2 = row['team2']+" "+'Vs'+" "+ row['team1']
    row_match_ID[key1] = row['match ID']
    row_match_ID[key2] = row['match ID']
```

```python
row_match_ID["U.A.E. Vs Netherlands"]
```

```python
df_batt.head()
```

```python
df_batt["match_ID"] = df_batt['match'].map(row_match_ID)
df_batt.head()
```

```python
df_batt.to_csv("t20_csv_files/batting_match_summary.csv", index = False)
```

```python
with open("t20_json_files/t20_wc_bowling_summary.json") as p:
    h = json.load(p)

    records_bowl = []
    for rek in h:
        records_bowl.extend(rek["bowlingSummary"])
```

```python
df_bowl = pd.DataFrame(records_bowl)
```

```python
df_bowl.head()
```

```python
df_bowl['match_ID'] = df_bowl['match'].map(row_match_ID)
df_bowl.head()
```

```python
df_bowl.to_csv("t20_csv_files/bowling_match_summary.csv", index = False)
```
